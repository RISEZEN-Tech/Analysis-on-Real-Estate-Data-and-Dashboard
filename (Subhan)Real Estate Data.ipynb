```python
# Step 2: Define a function to convert currency strings to numeric rupees
def convert_to_rupees(value):
    if pd.isna(value):
        return None
    value = str(value).lower().replace(',', '').strip()
    
    try:
        if 'Cr' in value or 'crore' in value:
            num = float(value.replace('rs', '').replace('Cr', '').replace('crore', '').strip())
            return int(num * 1e7)
        elif 'Lac' in value or 'lakh' in value:
            num = float(value.replace('rs', '').replace('Lac', '').replace('lakh', '').strip())
            return int(num * 1e5)
        elif 'rs' in value:
            num = float(value.replace('rs', '').strip())
            return int(num)
        else:
            return int(float(value))  # assume already in rupees
    except:
        return None  # if conversion fails

# Step 3: Apply it to the correct column (replace 'Price' with your actual column name)
df_drop['Price_numeric'] = df_drop['Amount(in rupees)'].apply(convert_to_rupees)

# Step 4 (optional): Save the updated dataset
df_drop.to_csv(r'C:\Users\PMLS\Downloads\lab_solution\house_prices.csv', index=False)

# Step 5 (optional): Show a few rows
print(df_drop[['Price_numeric', 'Amount(in rupees)']].head())
```

```python
import re

def convert_amount_regex(amount_str):
    if pd.isna(amount_str) or amount_str == '':
        return None
    
    amount_str = str(amount_str).strip().lower()
    
    # Pattern to match numbers with cr/lac suffixes
    pattern = r'([\d.]+)\s*(cr|lac|lakh)'
    match = re.search(pattern, amount_str)
    
    if match:
        number = float(match.group(1))
        unit = match.group(2)
        
        if unit == 'Cr':
            return number * 10000000
        else:  # lac or lakh
            return number * 100000
    else:
        # Try to parse as plain number
        try:
            return float(amount_str)
        except:
            return None

df_drop['Amount(in rupees)'] = df_drop['Amount(in rupees)'].apply(convert_amount_regex)
df_drop
```
