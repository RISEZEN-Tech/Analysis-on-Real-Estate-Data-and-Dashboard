```python
import pandas as pd
df = pd.read_csv(r"C:\Users\PMLS\Downloads\lab_solution\updated_house_prices.csv")
df.head()
```

```python
df["Price"] = pd.to_numeric(df["Price"]).astype(int)
df.head()
```

```python
df["Carpet Area"] = pd.to_numeric(df["Carpet Area"]).astype(int)
df.head()
```

```python
median_furnish = df["Furnishing"].fillna(method = "bfill")
df["Furnishing"] = df["Furnishing"].fillna(median_furnish)

df["Furnishing"] = pd.to_numeric(df["Furnishing"]).astype(int)
df.head()
```

```python
df["Super Area"] = pd.to_numeric(df["Super Area"]).astype(int)
fill_tran = df["Transaction"].fillna(method = 'ffill')
df["Transaction"] = df["Transaction"].fillna(fill_tran)
fill_bath = df["Bathroom"].fillna(method = 'ffill')
df["Bathroom"] = df["Bathroom"].fillna(fill_bath)
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, KBinsDiscretizer, LabelEncoder
le = LabelEncoder()
df["Bathroom"] = le.fit_transform(df["Bathroom"]).astype(str)
df["Transaction"] = le.fit_transform(df["Transaction"]).astype(str)
df["Bathroom"] = df["Bathroom"].astype(int)
df["Transaction"] = df["Transaction"].astype(int)
df.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
# Features (X) and target (y)
X = df[['Carpet Area', 'Transaction', 'Furnishing', 'overlooking',	'Society',	'Bathroom', 'Ownership', 'Super Area']]  # Only numeric columns
y = df['Price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)

# Train model
model = DecisionTreeRegressor(random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Evaluation
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

# Print metrics
print(f"Mean Absolute Error (MAE): ₹{mae:,.2f}")
print(f"Root Mean Squared Error (RMSE): ₹{rmse:,.2f}")
print(f"R² Score: {r2:.4f}")
```

```python
import pandas as pd
import plotly.express as px
from dash import Dash, html, dcc, Input, Output

# Load your processed data
df = pd.read_csv(r"C:\Users\PMLS\Downloads\new_house_price_dataset.csv")  # Replace with your real file
# Make sure 'Price' is numeric
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

# Features for dashboard
features = ['Carpet Area', 'Transaction', 'Furnishing', 'overlooking', 'Society', 'Bathroom', 'Ownership', 'Super Area']

# Initialize Dash app
app = Dash(__name__)
server = app.server  # For deployment if needed

# Layout
app.layout = html.Div([
    html.H1("🏠 House Price Dashboard", style={'textAlign': 'center'}),

    html.Div([
        html.Label("Select Feature to Compare with Price:"),
        dcc.Dropdown(
            id='feature-dropdown',
            options=[{'label': feature, 'value': feature} for feature in features],
            value='Carpet Area',
            style={'width': '60%'}
        )
    ], style={'padding': '20px'}),

    dcc.Graph(id='scatter-plot'),

    html.Div(id='stats-output', style={'padding': '20px', 'fontSize': '18px'})
])

# Callback to update scatter plot
@app.callback(
    Output('scatter-plot', 'figure'),
    Output('stats-output', 'children'),
    Input('feature-dropdown', 'value')
)
def update_graph(selected_feature):
    fig = px.scatter(
        df, x=selected_feature, y='Price',
        title=f"Price vs {selected_feature}",
        labels={selected_feature: selected_feature, 'Price': 'Price (₹)'},
        opacity=0.6,
        trendline="ols",

    )

    avg_price = df.groupby(selected_feature)['Price'].mean().mean()
    return fig, f"Average Price based on selected feature: ₹{avg_price:,.2f}"

# Run the app
if __name__ == '__main__':
    app.run(debug=True)

```
